Anaconda es una plataforma usada por data scientist y en machine learning. Multiplataforma (windows, Linux, OSx) con las herramientas necesarias para empezar a analizar información, con librerias como tnsorflow, scikit-learn para machine learnign y pandas, numpy para analisis.

**Descargar desde: [anaconda.com](https://www.anaconda.com/distribution/)**

Se puede tener un ambiente para analisis de imagenes y otro para enternamiento neuronal por ejemplo.
